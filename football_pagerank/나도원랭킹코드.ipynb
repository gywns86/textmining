```python
import pandas as pd
import xlsxwriter

data = pd.read_excel('data/fbs_data_정규시즌.xlsx')

#loser = data[5].dropna()
#winner = data[6].dropna()
game_result = []
#print(loser)
for a in range(0, len(data)):
    b = list(data.loc[a])
    game_result.append(b)
    
#print(game_result)
#list(data.loc[0])
game_result
```

```python
import numpy as np

ID_list = []
for result_list in game_result:
    for a in result_list:
        ID_list.append(a)
        final_ID = list(set(ID_list))
dic_list = {}
key_value = 0

for b in final_ID:
    dic_list[b] = key_value
    key_value += 1

matrix_lists = []
for a in range(len(final_ID)):
    for b in range(len(final_ID)):
        matrix_lists.append([a,b])
list_inverts = []
for result_list in game_result:
    invert_a= [dic_list[result_list[0]], dic_list[result_list[1]]]
    list_inverts.append(invert_a)

matrix = [[0.0 for col in range(len(final_ID))] for row in range(len(final_ID))]
matrix = np.array(matrix)

for matrix_list in matrix_lists:
    for list_invert in list_inverts:
        if matrix_list == list_invert:
            matrix[list_invert[0],list_invert[1]] += 1
matrix_list = matrix.tolist()

#PageRank 값 산출 코드

DF = 0.15 #덤핑팩트
rows_sum =matrix.sum(axis =1)
colomn_sum=matrix.sum(axis = 0)
row = 0
a=[]
for row_sum in rows_sum:
    if row_sum == 0:
        b = matrix[row]+1.0/len(final_ID)
        a.append(b.tolist())
    else:
        b = matrix[row]/row_sum
        a.append(b.tolist())
    row+=1
first_matrix = np.array(a)
final_matrix = (1-DF)*first_matrix+DF/len(final_ID)
first_value = matrix.sum(axis=0)/matrix.sum()
PR_value = np.dot(first_value, final_matrix)
row = 0
while row < 10000:
    row+=1
    PR_value = np.dot(PR_value, final_matrix)
    
print(PR_value)
#승률계산
win_ratio=[]
for a in range(0, len(final_ID)):
    win = colomn_sum[a]/(rows_sum[a]+colomn_sum[a])
    win_ratio.append(win)

modi_rank_value =[]
for b in range(0,len(final_ID)):
    modi_rank_value.append((PR_value[b])*win_ratio[b])

#PageRank 순위 코드
seq = sorted(PR_value, reverse = True)
index = np.array([seq.index(v) for v in PR_value]) + 1
index_final = index.tolist()
PR_RANK = {}
key_value = 0
for b in final_ID:
    PR_RANK[b] = index_final[key_value]
    key_value += 1

print(PR_RANK) #최종 PR_RANK 딕션러리 형태로 랭킹을 산출하였음

#랭킹점수산출코드
index_float = []
index_round = []
score_list = []
score = {}

for a in index:
    b = a+0.00
    index_float.append(b)
index = np.array(index_float)
score_cal = ((100-(index/len(final_ID)*100.0))+10)*10
score_cal = score_cal.tolist()
for a in score_cal:
    index_round.append(int(round(a)))

key_value_1 = 0
for a in final_ID:
    score[a] = index_round[key_value_1]
    key_value_1 +=1

print(score)#최종 socre 딕션러리 형태로 랭킹점수를 산출하였음
```

```python
import numpy as np

def softmax(x):
    e_x = np.exp(x - np.max(x))  # 각 요소에서 입력 벡터의 최댓값을 빼줌으로써 오버플로우 방지
    return e_x / e_x.sum(axis=0)

# 예시로 사용할 입력 벡터
x = np.array([2.0, 1.0, 0.1])
output = softmax(x)
print("소프트맥스 출력:", output)
```
